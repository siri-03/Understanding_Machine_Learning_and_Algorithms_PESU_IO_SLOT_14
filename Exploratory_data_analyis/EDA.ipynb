```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
data_frame=pd.DataFrame(data=iris.data,columns=iris.feature_names)
data_frame["target"]=iris.target
data_frame.head()
```

```python
data_frame.shape
```

```python
data_frame.describe()
```

```python
data_frame.info()
```

```python
data_frame.target.value_counts()
```

```python
sns.catplot(x="target",y="sepal length (cm)",data=data_frame,kind="box",aspect=1.5)
plt.title("Boxplot for target and sepal length")
plt.show()
```

```python
sns.catplot(x="target",y="petal length (cm)",data=data_frame,kind="box",aspect=1.5)
plt.title("Boxplot for target and petal length")
plt.show()
```

From the above two boxplots, we observe that the sepal length and petal length median of type 2 is greater than the other types(0 & 1)

```python
sns.scatterplot(x="sepal length (cm)",y="petal length (cm)",hue="target",data=data_frame,palette="Dark2",s=80)
```

**INFERENCE:**From the scatterplot we can infer that type 2 iris plants have a higher sepal length and petal length while type 0 has lower sepal as well as petal length
